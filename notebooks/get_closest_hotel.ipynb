```python
from DBEngine import db_engine
```

```python
def get_closest_hotels(lat, lon, n):
  def command(cursor): 
    cursor.execute("select * from hotels order by ST_Distance(ST_SetSRID(ST_Point(%(lon)s,%(lat)s),4326), hotels.point) limit %(n)s", {
      'lat': lat,
      'lon': lon,
      'n': n
    })

    return cursor.fetchall()
    
  return db_engine.execute(command)
```

```python
get_closest_hotels(27.494688,-82.575622,5)
```
